```python
import pandas as pd
```

```python
data_url = "https://raw.githubusercontent.com/fenago/tf/main/Chapter6-Regularization_and_Hyperparameter_Tuning/dataset/shuttle.trn"

data = pd.read_table(data_url, header=None, sep=' ')
data.head()
```

```python
data.isnull().sum()
```

```python
y = data.pop(9)
X = data.copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, stratify=y)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
```

```python
tf.random.set_seed(8)
```

```python
model = tf.keras.Sequential()
```

```python
fc1 = Dense(512, input_shape=(9,), activation='relu')
```

```python
fc2 = Dense(512, activation='relu')
fc3 = Dense(128, activation='relu')
fc4 = Dense(128, activation='relu')
```

```python
fc5 = Dense(8, activation='softmax')
```

```python
model.add(fc1)
model.add(fc2)
model.add(fc3)
model.add(fc4)
model.add(fc5)
```

```python
model.summary()
```

```python
loss = tf.keras.losses.SparseCategoricalCrossentropy()
```

```python
optimizer = tf.keras.optimizers.Adam(0.001)
```

```python
model.compile(optimizer=optimizer, loss=loss, \
                  metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=5, validation_split=0.2)
```

```python
reg_fc1 = Dense(512, input_shape=(9,), activation='relu', \
                    kernel_regularizer=tf.keras.regularizers\
                                         .l2(l=0.1))
reg_fc2 = Dense(512, activation='relu', \
                    kernel_regularizer=tf.keras.regularizers\
                                         .l2(l=0.1))
reg_fc3 = Dense(128, activation='relu', \
                    kernel_regularizer=tf.keras.regularizers\
                                         .l2(l=0.1))
reg_fc4 = Dense(128, activation='relu', \
                    kernel_regularizer=tf.keras.regularizers\
                                         .l2(l=0.1))
reg_fc5 = Dense(8, activation='softmax')
```

```python
model2 = tf.keras.Sequential()
model2.add(reg_fc1)
model2.add(reg_fc2)
model2.add(reg_fc3)
model2.add(reg_fc4)
model2.add(reg_fc5)
```

```python
model2.summary()
```

```python
model2.compile(optimizer=optimizer, loss=loss, \
                   metrics=['accuracy'])
```

```python
model2.fit(X_train, y_train, epochs=5, validation_split=0.2)
```
